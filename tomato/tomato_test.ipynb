```python
import os
from PIL import Image

# 이미지가 저장된 폴더 경로
folder_path = 'D:/AIHub/image/test'

# 폴더 내의 모든 파일 이름 가져오기
filenames = os.listdir(folder_path)

for filename in filenames:
    # 파일이 이미지인 경우에만 처리
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # 이미지 파일의 전체 경로
        image_path = os.path.join(folder_path, filename)

        # 이미지 파일 열기
        img = Image.open(image_path)

        # 이미지를 135도 회전
        img_rotated = img.rotate(135)
        

        # 회전된 이미지의 저장 경로 (원본 이미지 파일 이름에 '_rotated'를 추가)
        save_path = os.path.join("D:/AIHub/image/test/rotated", os.path.splitext(filename)[0] + '_rotated135' + os.path.splitext(filename)[1])

        # 회전된 이미지 저장
        img_rotated.save(save_path)
#45,90도의 회전도 진행하였음
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf

# 이미지 데이터를 저장할 리스트
img_data = []

for filename in os.listdir("D:/AIHub/image/test/rotated"):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        img_path = os.path.join("D:/AIHub/image/test/rotated", filename)
        img = Image.open(img_path)
        img = img.resize((128, 128))
        img_array = np.array(img) / 255.0
        img_data.append(img_array)

# 레이블 데이터 생성 
labels = np.random.randint(0, 2, len(img_data))

# 학습 데이터셋과 테스트 데이터셋으로 분리
X_train, X_test, Y_train, Y_test = train_test_split(img_data, labels, test_size=0.2, random_state=42)

# TensorFlow 데이터셋 구성
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, Y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, Y_test))

# 데이터셋 셔플 및 배치 사이즈 설정
train_dataset = train_dataset.shuffle(buffer_size=1000).batch(batch_size=32)
test_dataset = test_dataset.batch(batch_size=32)
```

```python
from tensorflow.keras import layers

# CNN 모델 구성
model = tf.keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 학습
model.fit(train_dataset, epochs=10, validation_data=test_dataset)
```

```python
test_loss, test_accuracy = model.evaluate(test_dataset)

print('테스트 데이터셋에 대한 손실값: ', test_loss)
print('테스트 데이터셋에 대한 정확도: ', test_accuracy)
```
