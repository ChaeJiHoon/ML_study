```python
import cv2
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np

# U-Net 모델 불러오기
model = load_model('unet_model_truss2.h5', custom_objects={'ConvBlock': ConvBlock})

# 이미지 불러오기
input_image = cv2.imread("D:\\1234.png")  # 파일 경로에 따라 'your_image.jpg'를 적절히 수정하세요.

if input_image is None:
    print('Image not loaded.')

# 이미지 전처리
input_image = cv2.resize(input_image, (272,480))  # U-Net에 맞게 이미지 크기 조절
input_image = input_image / 255.0  # 정규화
input_image = input_image[np.newaxis, ...]  # 배치 차원 추가

# U-Net을 이용한 이미지 세그멘테이션
prediction = model.predict(input_image)
segmented_image = np.argmax(prediction, axis=-1)  # 가장 높은 확률을 가진 클래스 선택

# 각 클래스를 특정 색상으로 표시하기 위한 색상 맵 생성
color_map = np.array([[0, 0, 0],   # 라벨 0: 검은색
                      [0, 0, 255], # 라벨 1: 빨간색
                      [0, 255, 0], # 라벨 2: 녹색
                      [255, 0, 0]], # 라벨 3: 파란색
                     dtype='uint8')

# 세그멘테이션 맵을 색상 이미지로 변환
segmented_image_color = color_map[segmented_image.squeeze()]

# 결과 출력
plt.imshow(segmented_image_color)
plt.show()
```
